```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset = pd.read_csv("../../data/intermediante_files/dataset_corrected.csv")
dataset.date = pd.to_datetime(dataset.date)
```

```python
dataset.shape
```

# Step 1: Make weekly dataset

```python
agg_dict = {
    "unit_sales":  "sum",
    "unit_sales_corrected": "sum",
    "onpromotion": "mean",
    "family": "first",
    "class": "first",
    "perishable": "first",
    "type": "first",
    "cluster": "first"
}

res = []
idx = 1
for store_nbr in dataset.store_nbr.unique():
    print(store_nbr, end="\r")
    dataset_w = dataset.loc[dataset.store_nbr == store_nbr]\
            .set_index("date")\
            .groupby([pd.Grouper(freq="W"), "store_nbr", "item_nbr"], as_index=True).agg(agg_dict)\
            .reset_index()#.to_csv('../data/weekly_chunk_{}.csv'.format(idx), index=False)
    idx += 1
    res.append(dataset_w)
```

```python
new_sd = pd.concat(res)
```

```python
new_sd.to_csv("../../data/intermediante_files/weekly_corrected_dataset.csv", index=False)
```
